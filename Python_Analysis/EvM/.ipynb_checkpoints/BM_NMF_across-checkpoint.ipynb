```python
def get_Region_data_con(data_con):
    CIRC_AREAS_FILEPATH = 'X:\\4 e-Lab\e-Lab shared code\Softwares\Connectogram\circ_areas.xlsx'
    atlas = pd.read_excel(CIRC_AREAS_FILEPATH, sheet_name='atlas')
    for subregion in np.unique(data_con[['StimA', 'ChanA']]):
        region = atlas.loc[atlas.Abbreviation == subregion, 'Area'].values
        if len(region) > 0:
            data_con.loc[data_con.StimA == subregion, 'StimR'] = region[0]
            data_con.loc[data_con.ChanA == subregion, 'ChanR'] = region[0]
        else:
            print(subregion)
            data_con.loc[data_con.StimA == subregion, 'StimR'] = 'U'
            data_con.loc[data_con.ChanA == subregion, 'ChanR'] = 'U'
    return data_con
```

```python

df = data_con_all

# Create pivot tables for each NREM value (-1, 0, 1)
pivot_tables = {}
for nrem_value in [-1, 0, 1]:
    pivot_table = df[df['REM'] == nrem_value].pivot_table(
        index='StimR',
        columns='ChanR',
        aggfunc='count',
        values ='d',
        fill_value=0
    )
    pivot_tables[nrem_value] = pivot_table

# Calculate percentages
total_counts = df.pivot_table(
    index='StimR',
    columns='ChanR',
    aggfunc='count',
    values ='d',
    fill_value=0
)

percentage_matrices = {}
for nrem_value, pivot_table in pivot_tables.items():
    percentage_matrix = (pivot_table / total_counts) * 100
    percentage_matrices[nrem_value] = percentage_matrix

# Plot the connectivity matrices as heatmaps
for nrem_value, percentage_matrix in percentage_matrices.items():
    plt.figure(figsize=(8, 6))
    plt.title(f'REM == {nrem_value}')
    sns.heatmap(percentage_matrix, annot=True, cmap="autumn", fmt=".2f", linewidths=.5)
    plt.show()
```

```python
# Create pivot tables for each NREM value (-1, 0, 1)
pivot_tables = {}
for nrem_value in [-1, 0, 1]:
    pivot_table = df[df['REM'] == nrem_value].pivot_table(
        index='StimR',
        columns='ChanR',
        aggfunc='count',
        values ='d',
        fill_value=0
    )
    pivot_tables[nrem_value] = pivot_table

# Calculate percentages
total_counts = df.pivot_table(
    index='StimR',
    columns='ChanR',
    aggfunc='count',
    values ='d',
    fill_value=0
)

percentage_matrices = {}
for nrem_value, pivot_table in pivot_tables.items():
    percentage_matrix = np.round((pivot_table / total_counts) * 100,1)
    percentage_matrix = percentage_matrix.fillna(0)
    percentage_matrices[nrem_value] = percentage_matrix

# Plot the connectivity matrices as heatmaps
for nrem_value, percentage_matrix in percentage_matrices.items():
    plt.figure(figsize=(8, 6))
    plt.title(f'REM == {nrem_value}')
    
    # Extract the corresponding pivot_table for raw counts
    raw_counts_matrix = pivot_tables[nrem_value]
    
    # Create a new matrix for annotations containing both percentage and raw count
    annot_matrix = percentage_matrices[nrem_value].astype(str) + "\n(" + raw_counts_matrix.astype(str) + ")"
    
    sns.heatmap(percentage_matrix, annot=annot_matrix, cmap="hot", fmt="", linewidths=.5)
    plt.show()# Plot the connectivity matrices as heatmaps

```
