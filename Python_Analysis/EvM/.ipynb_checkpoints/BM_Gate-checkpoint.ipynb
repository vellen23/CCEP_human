```python

# Assuming con_trial is your pandas DataFrame
con_trial['Con_ID'] = con_trial.groupby(['Stim', 'Chan']).ngroup()

# Calculate the proportions for each Con_ID
con_id_proportions = con_trial.groupby('Con_ID')['Sig'].mean()

# Filter out Con_IDs with a proportion of 0
non_zero_con_ids = con_id_proportions[con_id_proportions > 0].index

# Initialize an empty list to store arrays
M = []

for c in non_zero_con_ids:
    array = con_trial.loc[con_trial.Con_ID == c, 'Sig'].values
    M.append(array)
    

# Find the maximum length among all arrays
max_len = max(len(a) for a in M)
min_len = min(len(a) for a in M)

# Pad each array to the maximum length and create a 2D numpy array
M = np.array([np.pad(a, (0, max_len - len(a)), 'constant') for a in M])
M = M[:,:min_len]
# If you want the proportions of 1's in each row
proportions = np.mean(M, axis=1)

 #Sort the matrix rows by descending proportions
# Get the indices that would sort the proportions array
sort_indices = np.argsort(-proportions)

# Use these indices to sort M and proportions
sorted_M = M[sort_indices]
sorted_proportions = proportions[sort_indices]

```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

def run_length_analysis(array):
    diffs = np.diff(array)
    run_ends = np.where(diffs != 0)[0]
    run_lengths = np.diff(np.concatenate(([0], run_ends + 1, [len(array)])))
    return run_lengths

def wald_wolfowitz_test(array):
    runs = len(np.where(np.diff(array) != 0)[0]) + 1
    num_zeros = np.sum(array == 0)
    num_ones = np.sum(array == 1)
    expected_runs = ((2 * num_zeros * num_ones) / (num_zeros + num_ones)) + 1
    variance_runs = (expected_runs - 1) * (expected_runs - 2) / (num_zeros + num_ones - 1)
    z = (runs - expected_runs) / np.sqrt(variance_runs)
    p_value = stats.norm.sf(abs(z))  # two-tailed test
    return z, p_value

# Calculate metrics for each Con_ID and create a summary DataFrame
summary_data = []

for c in non_zero_con_ids:
    array = con_trial.loc[con_trial.Con_ID == c, 'Sig'].values
    p = np.mean(array)
    run_lengths = run_length_analysis(array)
    rla_var = np.var(run_lengths)
    wald_z, wald_p = wald_wolfowitz_test(array)
    summary_data.append([c, p, rla_var, wald_z, wald_p])

summary_df = pd.DataFrame(summary_data, columns=['Con_ID', 'P', 'RLA_var', 'Wald_z', 'Wald_p'])

summary_df['Sig'] = summary_df['Wald_p']<0.05
```
